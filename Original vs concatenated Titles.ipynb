# Checking combination and base

## Replace parenthesis with empty string and other preprocessing

```python
#df['Token'] = df['Token'].str.replace(r'[()]', '', regex=True).copy()
#df['Title'] = df['Title'].str.replace(r'[()]', '', regex=True).copy()
df['Title'] = df['Title'].str.replace(r'\s{2,}', ' ', regex=True)
df['Title'] = df['Title'].str.replace('\xa0', ' ', regex=True)
#df['Title'] = df['Title'].str.strip()
#df['Token'] = df['Token'].str.replace(r'\s{2,}', ' ', regex=True)
#df['Token'] = df['Token'].str.strip()
```

## Combine tokens

```python
combined_tokens_df = df.groupby('Record Number', sort=False)['Token'].apply(' '.join).reset_index()
combined_tokens_df
```

## Merge dataframes, replace non-breaking spaces, extra spaces

```python
merged_df = pd.merge(df[['Record Number', 'Title']], combined_tokens_df, on = 'Record Number')
merged_df = merged_df.drop_duplicates()
```

```python
merged_df['match'] = merged_df['Title']==merged_df['Token']
merged_df['match'].sum()
```

```python
unequal_titles_df = merged_df[merged_df['match']==False].reset_index(drop=True)[['Record Number', 'Title', 'Token']]
unequal_titles_df
```

```python
unequal_titles_df['longer_string'] = unequal_titles_df.apply(
    lambda row: 'Title' if len(row['Title']) > len(row['Token']) 
    else ('Token' if len(row['Title']) < len(row['Token']) else 'equal'),
    axis=1
)
```

```python
unequal_titles_df['longer_string'].value_counts()
```
