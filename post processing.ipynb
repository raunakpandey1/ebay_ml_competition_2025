# Remove puncuation only aspect names
# Remove spaces around '.' in each aspect value

```python
import string
import re
```

```python
# Use regex to match only-punctuation strings
df_no_punct = df[~df['Aspect Value'].str.fullmatch(rf"[{re.escape(string.punctuation)}]+")]
```

```python
df_no_punct
```

```python
df_removed_spaces_around_period = df_no_punct.copy()
```

```python
#only period
#df_removed_spaces_around_period['Aspect Value'] = df_removed_spaces_around_period['Aspect Value'].str.replace(r'\s*\.\s*', '.', regex=True)

#all punctuations
df_removed_spaces_around_period['Aspect Value'] = df_removed_spaces_around_period['Aspect Value'].str.replace(
    rf'\s*([{re.escape(string.punctuation)}])\s*', r'\1', regex=True
)

```
